```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/synthetic_dataset.xlsx'
data = pd.read_excel(file_path)

# Round and convert fault_location to integers
if 'fault_location' in data.columns:
    data['fault_location'] = data['fault_location'].round().astype(int)

# Ensure fault_type is treated as categorical
if 'fault_type' in data.columns:
    data['fault_type'] = data['fault_type'].astype(str)

# Define features and targets
X = data.drop(columns=['fault_type', 'fault_location'])
y = data[['fault_type', 'fault_location']]

# Split the data into training and testing sets
# Stratify using `fault_location` or `fault_type`
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y['fault_location'], test_size=0.3, random_state=89
)

# Separate fault type and fault location after splitting
y_train_type = y_train['fault_type']
y_train_loc = y_train['fault_location']
y_test_type = y_test['fault_type']
y_test_loc = y_test['fault_location']

# Scale features for KNN
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Function to evaluate KNN and plot accuracy
def evaluate_knn(X_train, X_test, y_train, y_test, label):
    training_accuracy = []
    test_accuracy = []
    neighbors_settings = range(1,8)

    for n_neighbors in neighbors_settings:
        # Build the model
        knn = KNeighborsClassifier(n_neighbors=n_neighbors)
        knn.fit(X_train, y_train)
        # Record training and test set accuracy
        training_accuracy.append(knn.score(X_train, y_train))
        test_accuracy.append(knn.score(X_test, y_test))

    # Plot accuracy
    plt.figure()
    plt.plot(neighbors_settings, training_accuracy, label="Training accuracy")
    plt.plot(neighbors_settings, test_accuracy, label="Test accuracy")
    plt.ylabel("Accuracy")
    plt.xlabel("n_neighbors")
    plt.legend()
    plt.title(f"KNN Accuracy for {label}")
    plt.show()

    # Report best n_neighbors
    best_n = neighbors_settings[test_accuracy.index(max(test_accuracy))]
    print(f"Best n_neighbors for {label}: {best_n}")
    print(f"Training Accuracy: {max(training_accuracy):.3f}")
    print(f"Testing Accuracy: {max(test_accuracy):.3f}")

# Evaluate for Fault Type
evaluate_knn(X_train, X_test, y_train_type, y_test_type, label="Fault Type")

# Evaluate for Fault Location
evaluate_knn(X_train, X_test, y_train_loc, y_test_loc, label="Fault Location")
```

synthetic dataset generation code

```python
import pandas as pd
import numpy as np


# Function to augment data using interpolation, noise addition, and synthetic generation
def generate_data(df, num_samples):
    augmented_data = df.copy()

    # 1. Interpolation: Generating intermediate rows by interpolating existing rows
    interp_samples = int(num_samples * 0.4)  # 40% of new samples
    interp_indices = np.random.choice(len(df) - 1, interp_samples)
    interp_data = pd.DataFrame([
        df.iloc[idx] + (df.iloc[idx + 1] - df.iloc[idx]) * np.random.random()
        for idx in interp_indices
    ], columns=df.columns)
    augmented_data = pd.concat([augmented_data, interp_data], ignore_index=True)

    # 2. Noise Addition: Adding Gaussian noise to randomly selected rows
    noise_samples = int(num_samples * 0.4)  # 40% of new samples
    noise_indices = np.random.choice(len(df), noise_samples)
    noise_std_dev = 0.05  # Standard deviation of noise as a fraction of data range
    noise_data = df.iloc[noise_indices].copy()
    for col in df.columns[:-2]:  # Exclude fault_type and fault_location from noise
        column_range = df[col].max() - df[col].min()
        noise_data[col] += np.random.normal(0, noise_std_dev * column_range, noise_samples)
    augmented_data = pd.concat([augmented_data, noise_data], ignore_index=True)

    # 3. Synthetic Data Generation: Sampling from distributions inferred from the data
    synth_samples = num_samples - len(augmented_data)  # Remaining samples
    synth_data = pd.DataFrame({
        col: np.random.normal(df[col].mean(), df[col].std(), synth_samples)
        if col != 'fault_type' and col != 'fault_location'
        else np.random.choice(df[col].unique(), synth_samples)
        for col in df.columns
    })
    augmented_data = pd.concat([augmented_data, synth_data], ignore_index=True)

    # Ensure fault_type and fault_location are integers
    augmented_data['fault_type'] = augmented_data['fault_type'].astype(int)
    augmented_data['fault_location'] = augmented_data['fault_location'].astype(int)

    # Shuffle the augmented data
    augmented_data = augmented_data.sample(frac=1).reset_index(drop=True)

    return augmented_data

# Load the original dataset
file_path = '/content/final_dataset.xlsx'
df = pd.read_excel(file_path)

# Generate 5000 samples
augmented_df = generate_data(df, 5000)

# Save the generated data to a new Excel file
output_path = '/content/synthetic_dataset_new.xlsx'
augmented_df.to_excel(output_path, index=False)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import seaborn as sns
# Load the dataset
file_path = '/content/synthetic_dataset.xlsx'
data = pd.read_excel(file_path)


# Define features and targets
X = data.drop(columns=['fault_type', 'fault_location'])
y = data[['fault_type', 'fault_location']]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y['fault_location'], test_size=0.3, random_state=89
)

# Separate fault type and fault location after splitting
y_train_type = y_train['fault_type']
y_train_loc = y_train['fault_location']
y_test_type = y_test['fault_type']
y_test_loc = y_test['fault_location']

# Scale features for KNN and XGBoost
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


# Function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, label):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=np.unique(y_true), yticklabels=np.unique(y_true))
    plt.title(f"Confusion Matrix for {label}")
    plt.ylabel("Actual")
    plt.xlabel("Predicted")
    plt.show()


# Function to evaluate KNN with hyperparameter tuning
def evaluate_knn(X_train, X_test, y_train, y_test, label):
    # Hyperparameter tuning using GridSearchCV
    param_grid = {'n_neighbors': range(1, 15), 'metric': ['euclidean', 'manhattan', 'minkowski']}
    grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='accuracy')
    grid.fit(X_train, y_train)

    best_k = grid.best_params_['n_neighbors']
    best_metric = grid.best_params_['metric']
    print(f"Best KNN parameters for {label}: {grid.best_params_}")
    print(f"Best CV accuracy for {label}: {grid.best_score_:.3f}")

    # Evaluate KNN with best parameters
    knn = KNeighborsClassifier(n_neighbors=best_k, metric=best_metric)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)

    # Error metrics
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    print(f"\n{label} - Error Metrics (KNN):")
    print(f"MSE: {mse:.3f}, RMSE: {rmse:.3f}, MAE: {mae:.3f}, R2: {r2:.3f}")

    # Plot error metrics
    plt.figure()
    plt.bar(['MSE', 'RMSE', 'MAE', 'R2'], [mse, rmse, mae, r2], color=['blue', 'orange', 'green', 'red'])
    plt.title(f"KNN Error Metrics for {label}")
    plt.ylabel("Metric Value")
    plt.show()

# Evaluate KNN for both fault_type and fault_location
evaluate_knn(X_train, X_test, y_train_type, y_test_type, label="Fault Type")
evaluate_knn(X_train, X_test, y_train_loc, y_test_loc, label="Fault Location")

# XGBoost Implementation
def evaluate_xgboost(X_train, X_test, y_train, y_test, label):
    xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss')
    xgb_model.fit(X_train, y_train)
    y_pred = xgb_model.predict(X_test)

    # Error metrics
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    print(f"\n{label} - Error Metrics (XGBoost):")
    print(f"MSE: {mse:.3f}, RMSE: {rmse:.3f}, MAE: {mae:.3f}, R2: {r2:.3f}")

    # Plot error metrics
    plt.figure()
    plt.bar(['MSE', 'RMSE', 'MAE', 'R2'], [mse, rmse, mae, r2], color=['blue', 'orange', 'green', 'red'])
    plt.title(f"XGBoost Error Metrics for {label}")
    plt.ylabel("Metric Value")
    plt.show()



# Evaluate XGBoost for both fault_type and fault_location
evaluate_xgboost(X_train, X_test, y_train_type, y_test_type, label="Fault Type")
evaluate_xgboost(X_train, X_test, y_train_loc, y_test_loc, label="Fault Location")


def evaluate_random_forest(X_train, X_test, y_train, y_test, label):
    # Initialize and train Random Forest Classifier
    rf_model = RandomForestClassifier(random_state=42)
    rf_model.fit(X_train, y_train)
    y_pred = rf_model.predict(X_test)

    # Error metrics
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    print(f"\n{label} - Error Metrics (Random Forest):")
    print(f"MSE: {mse:.3f}, RMSE: {rmse:.3f}, MAE: {mae:.3f}, R2: {r2:.3f}")

    # Plot error metrics
    plt.figure()
    plt.bar(['MSE', 'RMSE', 'MAE', 'R2'], [mse, rmse, mae, r2], color=['blue', 'orange', 'green', 'red'])
    plt.title(f"Random Forest Error Metrics for {label}")
    plt.ylabel("Metric Value")
    plt.show()

# Evaluate Random Forest for both fault_type and fault_location
evaluate_random_forest(X_train, X_test, y_train_type, y_test_type, label="Fault Type")
evaluate_random_forest(X_train, X_test, y_train_loc, y_test_loc, label="Fault Location")

```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
import seaborn as sns
# Load the dataset
file_path = '/content/synthetic_dataset_new.xlsx'
data = pd.read_excel(file_path)


# Define features and targets
X = data.drop(columns=['fault_type', 'fault_location'])
y = data[['fault_type', 'fault_location']]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y['fault_location'], test_size=0.3, random_state=89
)

# Separate fault type and fault location after splitting
y_train_type = y_train['fault_type']
y_train_loc = y_train['fault_location']
y_test_type = y_test['fault_type']
y_test_loc = y_test['fault_location']

# Scale features for KNN and XGBoost
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


# Function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, label):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=np.unique(y_true), yticklabels=np.unique(y_true))
    plt.title(f"Confusion Matrix for {label}")
    plt.ylabel("Actual")
    plt.xlabel("Predicted")
    plt.show()

def plot_training_testing_accuracy(train_acc, test_acc, label):
    plt.figure(figsize=(6, 4))
    plt.bar(['Training', 'Testing'], [train_acc, test_acc], color=['green', 'blue'])
    plt.ylim(0, 1)
    plt.ylabel('Accuracy')
    plt.title(f"Training vs Testing Accuracy for {label}")
    plt.show()

def evaluate_knn(X_train, X_test, y_train, y_test, label):
    param_grid = {
        'n_neighbors': [3, 5, 7, 9, 11],
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan', 'minkowski']
    }

    # GridSearchCV for hyperparameter tuning
    knn_model = KNeighborsClassifier()
    grid_search = GridSearchCV(estimator=knn_model, param_grid=param_grid, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
    grid_search.fit(X_train, y_train)

    # Get the best parameters
    best_params = grid_search.best_params_
    print(f"\nBest Hyperparameters for {label}: {best_params}")

    # Train the model with the best parameters
    tuned_knn_model = grid_search.best_estimator_
    y_pred_train = tuned_knn_model.predict(X_train)
    y_pred_test = tuned_knn_model.predict(X_test)

    # Training and testing accuracy
    train_acc = accuracy_score(y_train, y_pred_train)
    test_acc = accuracy_score(y_test, y_pred_test)
    print(f"Training Accuracy (Tuned KNN): {train_acc:.3f}, Testing Accuracy (Tuned KNN): {test_acc:.3f}")

    # Plot training and testing accuracy
    plot_training_testing_accuracy(train_acc, test_acc, label)

    # Error metrics
    mse = mean_squared_error(y_test, y_pred_test)
    mae = mean_absolute_error(y_test, y_pred_test)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred_test)
    # mape = calculate_mape(y_test, y_pred_test)

    print(f"\n{label} - Error Metrics (Tuned KNN):")
    print(f"MSE: {mse:.3f}, RMSE: {rmse:.3f}, MAE: {mae:.3f}, R2: {r2:.3f}")
    plt.figure()
    plt.bar(['MSE', 'RMSE', 'MAE', 'R2'], [mse, rmse, mae, r2], color=['blue', 'orange', 'green', 'red'])
    plt.title(f"Tuned KNN Error Metrics for {label}")
    plt.ylabel("Metric Value")
    plt.show()

    # Confusion matrix
    plot_confusion_matrix(y_test, y_pred_test, label)

print("\nEvaluating KNN:")
evaluate_knn(X_train, X_test, y_train_type, y_test_type, label="Fault Type")
evaluate_knn(X_train, X_test, y_train_loc, y_test_loc, label="Fault Location")
```
